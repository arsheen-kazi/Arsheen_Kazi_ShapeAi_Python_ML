```python
import numpy as np
import statistics
from statistics import StatisticsError
import matplotlib.pyplot as plt
import pylab 
import scipy.stats as stats


speed = [13, 15, 16, 16, 19, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]



x = statistics.mean(speed)
y = statistics.median(speed)
a = np.std(speed)
b = np.var(speed)
c = np.quantile(speed, [0,0.25,0.5,0.75,1])

print('Mean is ',x)
print('Median is',y)
try:
    z = statistics.mode(speed)
except StatisticsError:
    print ("No unique mode found")


print('Standard Deviation is',a)
print('Variance is',b)
print('Quantile is',c)
print('Minimum value is',np.min(speed))
print('Maximum value is',np.max(speed))
print("Mid range is ",speed[int(len(speed)/2)])


plt.style.use('ggplot')
plt.hist(speed, bins=10)
plt.show()



measurements = np.random.normal(loc = 20, scale = 5, size=100)   
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```
